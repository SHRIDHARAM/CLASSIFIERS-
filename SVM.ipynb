```python
import pandas as pd
import numpy as np
from sklearn.svm import SVC
```

# Q5. Part1

```python
columns = np.arange(5000)
```

```python
df_train = pd.read_csv('gisette_train.data', sep = " ", names= columns, index_col=False)
```

```python
df_train.head()
```

```python
df_train.shape
```

```python
train_label = pd.read_csv('gisette_train.labels', names = ['label'])
train_label.head()
```

```python
train_label.shape
```

```python
test_label = pd.read_csv('gisette_valid.labels', names = ['label'])
```

```python
test_label.shape
```

```python
df_test = pd.read_csv('gisette_valid.data', sep = " ", names= columns, index_col=False)
```

```python
df_test.head()
```

```python
df_test.shape
```

```python
from sklearn.svm import SVC
model = SVC(kernel="linear")
```

```python
model.fit(df_train, train_label)
```

```python
n = model.support_vectors_ 
print(len(n))
print(n)
```

```python
# for testing error
model.score(df_test, test_label)
```

```python
#for training error
model.score(df_train, train_label)
```

# Q5 Part 2

# RBF (a.k.a. Gaussian kernel; set γ = 0.001)

```python
model_gauss = SVC(kernel="rbf",gamma= 0.001)
```

```python
model_gauss.fit(df_train, train_label)
```

```python
m = model_gauss.support_vectors_ 
print(len(m))
print(m)
```

```python
#for training error
model_gauss.score(df_train, train_label)
```

```python
#for testing error
model_gauss.score(df_test, test_label)
```

# Polynomial kernel (set degree =2, coef0 = 1)

```python
model_poly = SVC(kernel="poly", degree=2, coef0 =1)
```

```python
model_poly.fit(df_train, train_label)
```

```python
o = model_poly.support_vectors_ 
print(len(o))
print(o)
```

```python
#for training error
model_poly.score(df_train, train_label)
```

```python
#for testing error
model_poly.score(df_test, test_label)
```
