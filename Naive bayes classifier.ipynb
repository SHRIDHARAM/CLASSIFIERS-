```python
import pandas as pd
import numpy as np
```

```python
columns = np.arange(1,58)
columns = np.append(columns,'labels')
print(columns)
```

```python
df = pd.read_csv('spambase.data', sep=",",names=columns,index_col=False)
```

```python
df.head()
```

from the documentation of data we found that each attribute is a continous type of attribute, hence we should use gaussian naive bayes, but we will check every model, bernoulli can binarize the data, and as these data are given so can use multinomial also, we will check for all of them, and get the best fitted model, but by intuation, gaussian can be applied directly


```python
Y_df = df['labels']
Y_df.head()
```

```python
X_df = df.drop('labels',axis=1)
X_df.head()
```

# Standardization

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_df)
```

```python
scaled_feat = scaler.transform(X_df)
```

```python
X_df_transformed = pd.DataFrame(scaled_feat, columns= df.columns[:-1])
X_df_transformed.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_df_transformed,df['labels'], test_size=0.20)
```

# Gaussian

```python
from sklearn.naive_bayes import GaussianNB 
classifier = GaussianNB()
classifier.fit(X_train,Y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(Y_test, y_pred))
```

```python
print(classification_report(Y_test, y_pred))
```
